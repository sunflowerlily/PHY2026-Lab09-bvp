```python
import numpy as np
import matplotlib.pyplot as plt
from lab1_core.P2_FDM_HeatRod.p2_fdm_heat import solve_temperature_profile
```

```python
L = 10.0
alpha = 0.1
T_env = 20.0
T_left = 100.0
T_right = 50.0

for N in [10, 50, 100]:
    x, T = solve_temperature_profile(L=L, alpha=alpha, T_env=T_env, T_left=T_left, T_right=T_right, N=N)
    print(f"N={N:3d}, T(0)={T[0]:.2f}, T(L)={T[-1]:.2f}, Tmin={T.min():.2f}, Tmax={T.max():.2f}")
```

```python
plt.figure(figsize=(7, 4))
for N in [10, 50, 100]:
    x, T = solve_temperature_profile(L=L, alpha=alpha, T_env=T_env, T_left=T_left, T_right=T_right, N=N)
    plt.plot(x, T, label=f"N={N}")

plt.title("FDM Temperature Profile")
plt.xlabel("x")
plt.ylabel("T")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
