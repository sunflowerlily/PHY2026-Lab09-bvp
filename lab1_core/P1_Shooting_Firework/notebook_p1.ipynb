```python
import numpy as np
import matplotlib.pyplot as plt
from lab1_core.P1_Shooting_Firework.p1_shooting import secant_search_for_v0, terminal_residual, integrate_trajectory
```

```python
T = 5.0
H = 40.0
gamma = 0.01
v0 = secant_search_for_v0(20.0, 80.0, T=T, H=H, gamma=gamma)
res = terminal_residual(v0, T=T, H=H, gamma=gamma)
print(f"v0 = {v0:.6f}")
print(f"|y(T)-H| = {abs(res):.3e}")
```

```python
t, y, v = integrate_trajectory(v0, T=T, gamma=gamma)
fig, ax = plt.subplots(1, 2, figsize=(10, 4))
ax[0].plot(t, y)
ax[0].set_title("Height vs Time")
ax[0].grid(True)
ax[1].plot(t, v)
ax[1].set_title("Velocity vs Time")
ax[1].grid(True)
plt.tight_layout()
plt.show()
```
